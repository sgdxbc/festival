```python
import pandas as pd
import seaborn as sns
sns.set_theme(style='ticks')

time_stat = pd.concat([
    pd.read_csv("./data/TimeStore.txt"),
    pd.read_csv("./data/TimeStore2.txt"),
    pd.read_csv("./data/TimeStore3.txt"),
])
# print(repair.info())
sns.relplot(
    data=time_stat, x='t', y='store', hue='protocol',
    kind='line',
    height=8, aspect=16 / 8,
)
```
