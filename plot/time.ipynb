```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(context='paper', style='ticks')

time_stat = pd.concat([
    pd.read_csv("./data/TimeStore.txt"),
    pd.read_csv("./data/TimeStore2.txt"),
    pd.read_csv("./data/TimeStore3.txt"),
])
# print(repair.info())
fig, axs = plt.subplots(2)
g = sns.lineplot(ax=axs[0], data=time_stat, x='t', y='store', hue='protocol', 
    hue_order=['entropy', 'entropyF', 'replicated'])
g.set(xlabel='', ylabel='Raw capacity (object size)', xlim=(0, 10 * 365 * 86400))
g.set_yticks(range(0, 5, 1))
g.set_ylabel('Raw capacity (object size)', labelpad=16)
sns.move_legend(g, 'upper right')
g = sns.lineplot(ax=axs[1], data=time_stat, x='t', y='repair', hue='protocol',
    hue_order=['entropy', 'entropyF', 'replicated'])
g.set(xlabel='Time (s)', ylabel='Repair traffic (object size)', xlim=(0, 10 * 365 * 86400))
g.set_yticks(range(0, 5, 1))
g.set_yscale('log')
g.legend([], [], frameon=False)
```
