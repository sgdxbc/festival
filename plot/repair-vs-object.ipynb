```python
import pandas as pd
import seaborn as sns
sns.set_theme(style='ticks')

repair = pd.read_csv("./data/RepairTraffic.txt")
# print(repair.info())
repair = repair[repair['protocol'] != 'entropy-0.1'].groupby(['protocol', '#object']).mean()
repair
sns.relplot(
    data=repair, x='#object', y='repair', hue='protocol',
    kind='line',
    height=8, aspect=16 / 8,
)
```
