```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.figure import figaspect
sns.set_theme(context='paper', style='ticks')

faulty = pd.read_csv("./data/FaultyLost.txt")
# print(repair.info())
fig, ax = plt.subplots(1, figsize=(4, 4 / (16 / 9)))
g = sns.lineplot(ax=ax, data=faulty, x='%faulty', y='#lost', hue='protocol',
    marker='o', err_style='bars')
g.set(ylim=(0, 1000), xlabel='Portion of faulty nodes', ylabel='Number of lost objects')
sns.move_legend(g, 'upper left')
```
