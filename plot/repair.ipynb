```python
import pandas as pd
import seaborn as sns
sns.set_theme(style='ticks')

repair = pd.read_csv("./data/RepairTraffic.txt")
# print(repair.info())
sns.relplot(
    data=repair[repair['AFR'] == 1], x='#object', y='repair', hue='protocol',
    kind='line',
    aspect=16 / 8,
)
```

```python
sns.relplot(
    data=repair[repair['#object'] == 1000], x='AFR', y='repair', hue='protocol',
    kind='line',
    aspect=16 / 8,
)
```
