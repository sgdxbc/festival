```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(context='paper', style='ticks')

repair = pd.read_csv("./data/RepairTraffic.txt")
# print(repair.info())

palette = dict(zip(['entropy-0', 'entropy-0.25', 'entropy-0.5', 'entropy-0.75', 'entropy-1'], sns.color_palette('magma', n_colors=5)))
palette['replicated'] = '0.6'
fig, axs = plt.subplots(2)
fig.tight_layout(pad=1.4)
g = sns.lineplot(ax=axs[0], data=repair[repair['AFR'] == 1], x='#object', y='repair', hue='protocol',
    marker='o', palette=palette)
g.set(xlabel='Number of objects', ylabel='Repair traffic (object size)', xscale='log', yscale='log')
sns.move_legend(g, 'upper left')
g = sns.lineplot(ax=axs[1], data=repair[repair['#object'] == 1000], x='AFR', y='repair', hue='protocol',
    marker='o', palette=palette)
g.set(xlabel='Annualized failure rate', ylabel='Repair traffic (object size)', xscale='log', yscale='log')
g.set_xticks([], minor=True)
g.set_xticks([0.5, 1., 2., 4])
g.set_xticklabels(['0.5', '1.', '2.', '4.'])
g.legend([], [], frameon=False)
```
